```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Datos de la tabla
data = {
    'Modelo': ['Hybrid RL-RF (2023)', 'ANN (2023)', 'Deep Q-Learning (2023)', 'SVM (2023)', 'Decision Trees [22]', 'GA-ARIMA', 'Reinforced RF'],
    'MAE': [3.45, 1.75, 0.99, 1.48, 0.48, 0.80, 1.28],
    'RMSE': [1.45, 3.48, 2.79, 0.76, 3.49, 3.75, 2.48],
    'RRSE': [0.78, 0.75, 0.45, 0.55, 1.48, 1.21, 0.54],
    'RAE': [2.49, 1.68, 1.67, 3.49, 3.47, 0.82, 0.91],
    'MSE': [2.50, 2.75, 3.75, 2.79, 2.49, 0.07, 0.55],
    'R²': [1.78, 3.12, 2.79, 1.70, 1.79, 0.54, 0.90]
}

df = pd.DataFrame(data)

# Métricas a graficar
metricas = ['MAE', 'RMSE', 'RRSE', 'RAE', 'MSE', 'R²']
n_metricas = len(metricas)
n_modelos = len(df['Modelo'])

# Ancho de las barras
ancho_barra = 0.12  # Ajustar un poco el ancho para mejor visualización con seaborn

# Posiciones de las barras para cada métrica
posiciones = np.arange(n_modelos)

# Usar la paleta de colores 'tab10' de seaborn
sns.set_palette('muted')
colores = sns.color_palette()

# Estilo de seaborn para una apariencia más profesional
sns.set_style("whitegrid")

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(14, 7))  # Aumentar un poco el tamaño para mejor visualización

# Graficar las barras agrupadas
for i, metrica in enumerate(metricas):
    offset = i * ancho_barra
    ax.bar(posiciones + offset, df[metrica], ancho_barra, label=metrica, color=colores[i], alpha=0.8)

# Etiquetas de los ejes y título
ax.set_xlabel('Modelo', fontsize=14)
ax.set_ylabel('Valor de la Métrica', fontsize=14)
ax.set_title('Comparación de Métricas por Modelo', fontsize=24, fontweight='bold')
ax.set_xticks(posiciones + (n_metricas * ancho_barra) / 2 - ancho_barra / 2)
ax.set_xticklabels(df['Modelo'], rotation=45, ha='right', fontsize=12)
ax.tick_params(axis='y', labelsize=12)

# Leyenda con un tamaño de fuente adecuado
ax.legend(title='Métrica', fontsize=12, loc='upper right')

# Guardar la figura en un archivo PNG
plt.savefig('comparacion_metricas_modelos.png', dpi=300, bbox_inches='tight')
# Mostrar la gráfica
plt.show()
```
